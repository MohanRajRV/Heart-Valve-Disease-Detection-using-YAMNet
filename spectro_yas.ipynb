```python
import os
import numpy as np
import librosa
import tensorflow as tf
import tensorflow_hub as hub
from scipy.signal import butter, filtfilt
from tensorflow.keras import layers, models
from sklearn.model_selection import train_test_split
```

```python
def create_linear_fragments(signal):

    spec = np.abs(librosa.stft(
        signal,
        n_fft=128,
        win_length=50,
        hop_length=20
    ))

    spec = np.log1p(spec)

    fragment_frames = 150
    fragments = []

    for start in range(0, spec.shape[1], fragment_frames):

        frag = spec[:, start:start + fragment_frames]

        if frag.shape[1] < fragment_frames:
            pad_width = fragment_frames - frag.shape[1]
            frag = np.pad(frag, ((0,0),(0,pad_width)), mode='constant')

        frag = tf.image.resize(
            frag[..., np.newaxis],
            (96, 64)
        ).numpy()

        frag = frag / np.max(frag)

        fragments.append(frag)

    return fragments
```

```python
from scipy.signal import butter, filtfilt
def bandpass_filter(signal, sr, low=25, high=600, order=10):

    nyquist = 0.5 * sr
    low = low / nyquist
    high = high / nyquist

    b, a = butter(order, [low, high], btype='band')

    filtered = filtfilt(b, a, signal)

    return filtered
```

```python
def preprocess_signal(path):

    try:
        signal, sr = librosa.load(path, sr=None)
    except Exception as e:

        print("Error loading:", path)
        print(e)
        return None

    print("Loaded:", path)

    if sr != 2000:
        signal = librosa.resample(signal, orig_sr=sr, target_sr=2000)
        sr = 2000

    signal = bandpass_filter(signal, sr)

    std = np.std(signal)

    if std == 0:
        print("STD ZERO:", path)
        return None

    signal = (signal - np.mean(signal)) / std

    duration = len(signal) / sr
    print("Duration:", duration)

    if duration < 1.6:
        print("Too short:", path)
        return None

    if duration > 10:
        signal = signal[:10 * sr]

    return signal.astype(np.float32)
```

```python
import os
import numpy as np

def load_dataset(dataset_path):

    X = []
    y = []

    class_names = ["N_New","MR_New","AS_New","MS_New","MVP_New"]

    for label, class_name in enumerate(class_names):

        folder = os.path.join(dataset_path, class_name)

        for file in os.listdir(folder):

            if file.endswith(".wav"):

                path = os.path.join(folder, file)

                signal = preprocess_signal(path)

                if signal is None:
                    continue


                fragments = create_linear_fragments(signal)


                for frag in fragments:
                    X.append(frag)
                    y.append(label)

                if len(X) % 200 == 0:
                    print("Processed:", len(X))

    return np.array(X), np.array(y)
```
