```python
import pywt
import tensorflow as tf

def create_scalogram_fragments(signal, sr=2000):


    scales = np.arange(1, 64)

    coef, _ = pywt.cwt(signal, scales, 'morl')

    scalogram = np.abs(coef)


    scalogram = scalogram / np.max(scalogram)
    scalogram = np.log(scalogram + 1e-6)

    fragment_frames = 150
    overlap = 0.5
    step = int(fragment_frames * (1 - overlap))

    fragments = []

    for start in range(0, scalogram.shape[1] - fragment_frames + 1, step):

        frag = scalogram[:, start:start + fragment_frames]

        frag = tf.image.resize(
            frag[..., np.newaxis],
            (96, 64)
        ).numpy()

        fragments.append(frag)

    return fragments
```

```python
from scipy.signal import butter, filtfilt

def bandpass_filter(signal, sr, low=20, high=400, order=4):

    nyquist = 0.5 * sr
    low = low / nyquist
    high = high / nyquist

    b, a = butter(order, [low, high], btype='band')

    filtered = filtfilt(b, a, signal)

    return filtered
```

```python
import librosa
import numpy as np

def preprocess_signal(path):

    try:
        signal, sr = librosa.load(path, sr=None)
    except Exception as e:
        print("Error loading:", path)
        print(e)
        return None
    print("Loaded:", path)

    # 🔹 Resample to 2000 Hz
    if sr != 2000:
        signal = librosa.resample(signal, orig_sr=sr, target_sr=2000)
        sr = 2000


    signal = bandpass_filter(signal, sr)


    signal = signal - np.mean(signal)

    std = np.std(signal)

    if std == 0:
        return None

    # 🔹 Z-score normalization
    signal = signal / std

    duration = len(signal) / sr

    # 🔹 Reject short signals
    if duration < 1.6:
        return None

    # 🔹 Crop long signals
    if duration > 10:
        signal = signal[:10 * sr]

    return signal.astype(np.float32)
```

```python

def load_dataset(dataset_path):

    X = []
    y = []

    class_names = ["N_New","MR_New","AS_New","MS_New","MVP_New"]

    for label, class_name in enumerate(class_names):

        folder = os.path.join(dataset_path, class_name)

        for file in os.listdir(folder):

            if file.endswith(".wav"):

                path = os.path.join(folder, file)

                signal = preprocess_signal(path)

                if signal is None:
                    continue

                fragments = create_scalogram_fragments(signal)



                for frag in fragments:
                    X.append(frag)
                    y.append(label)

                if len(X) % 200 == 0:
                    print("Processed:", len(X))


    return np.array(X, dtype=np.float32), np.array(y)
```

```python

train_path = "/content/drive/MyDrive/MINIPROJECT/yassen_filtered/train"
test_path  = "/content/drive/MyDrive/MINIPROJECT/yassen_filtered/test"




X_train, y_train = load_dataset(train_path)
X_test, y_test = load_dataset(test_path)

print("Train shape:", X_train.shape)
print("Test shape:", X_test.shape)
```

```python
from tensorflow.keras.optimizers import Adam

net.compile(
    optimizer = tf.keras.optimizers.Adam(learning_rate=0.0001),
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)

history = net.fit(
    X_train, y_train,
    epochs=40,
    batch_size=32,
)
```

```python
import numpy as np

y_pred_prob = net.predict(X_test)
y_pred = np.argmax(y_pred_prob, axis=1)
```

```python
from sklearn.metrics import (
    accuracy_score,
    precision_score,
    recall_score,
    f1_score
)
```

```python
accuracy = accuracy_score(y_test, y_pred)

precision = precision_score(y_test, y_pred, average='macro')
recall = recall_score(y_test, y_pred, average='macro')
f1 = f1_score(y_test, y_pred, average='macro')

print("Accuracy:", accuracy*100)
print("Precision:", precision)
print("UAR:", recall)
print("F1 Score:", f1)
```
