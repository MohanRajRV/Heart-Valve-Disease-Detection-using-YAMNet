```python
def create_logmel_fragments(signal, sr=2000):

    mel_spec = librosa.feature.melspectrogram(
        y=signal,
        sr=sr,
        n_fft=256,
        hop_length=64,
        n_mels=64
    )

    log_mel = librosa.power_to_db(mel_spec, ref=np.max)

    fragment_frames = 150
    hop = 4

    fragments = []

    for start in range(0, log_mel.shape[1], hop):

        frag = log_mel[:, start:start + fragment_frames]

        if frag.shape[1] < fragment_frames:
            pad_width = fragment_frames - frag.shape[1]
            frag = np.pad(frag, ((0,0),(0,pad_width)), mode='constant')

        frag = tf.image.resize(
            frag[..., np.newaxis],
            (96, 64)
        ).numpy()

        frag = (frag - np.min(frag)) / (np.max(frag) - np.min(frag) + 1e-8)

        fragments.append(frag)

    return fragments
```

```python
from scipy.signal import butter, filtfilt

def bandpass_filter(signal, sr, low=25, high=600, order=10):

    nyquist = 0.5 * sr
    low = low / nyquist
    high = high / nyquist

    b, a = butter(order, [low, high], btype='band')

    filtered = filtfilt(b, a, signal)

    return filtered
```

```python
def preprocess_signal(path):

    try:
        signal, sr = librosa.load(path, sr=None)
    except Exception as e:
        print("Error loading:", path)
        print(e)
        return None

    print("Loaded:", path)


    if sr != 2000:
        signal = librosa.resample(signal, orig_sr=sr, target_sr=2000)
        sr = 2000

    signal = bandpass_filter(signal, sr)


    std = np.std(signal)
    if std == 0:
        print("STD ZERO:", path)
        return None

    signal = (signal - np.mean(signal)) / std
    signal = signal + 0.002 * np.random.randn(len(signal))

    duration = len(signal) / sr
    print("Duration:", duration)

    if duration < 2:
        print("Too short:", path)
        return None

    max_length = 5 * sr

    if len(signal) > max_length:
        signal = signal[:max_length]

    return signal.astype(np.float32)
```

```python
import os
import numpy as np

def load_dataset(dataset_path):

    X = []   # list of fragment lists
    y = []   # labels

    class_names = ["N_New","MR_New","AS_New","MS_New","MVP_New"]

    for label, class_name in enumerate(class_names):

        folder = os.path.join(dataset_path, class_name)

        for file in os.listdir(folder):

            if file.endswith(".wav"):

                path = os.path.join(folder, file)

                signal = preprocess_signal(path)

                if signal is None:
                    continue

                fragments = create_logmel_fragments(signal)

                if len(fragments) == 0:
                    continue

                X.append(fragments)
                y.append(label)

    return np.array(X, dtype=object), np.array(y)
```

```python

train_path = "/content/drive/MyDrive/MINIPROJECT/yassen_filtered/train"
test_path  = "/content/drive/MyDrive/MINIPROJECT/yassen_filtered/test"


X_train, y_train = load_dataset(train_path)
X_test, y_test = load_dataset(test_path)

print("Train shape:", X_train.shape)
print("Test shape:", X_test.shape)
```

```python
from tensorflow.keras.optimizers import Adam

net.compile(
    optimizer = tf.keras.optimizers.Adam(learning_rate=0.0001),
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)

history = net.fit(
    X_train, y_train,
    epochs=100,
    batch_size=32,
)
```

```python
loss, acc = net.evaluate(X_test, y_test)
print("Test Accuracy:", acc)
```

```python
import numpy as np

y_pred_prob = net.predict(X_test)
y_pred = np.argmax(y_pred_prob, axis=1)
```

```python
from imblearn.metrics import specificity_score
from sklearn.metrics import (
    accuracy_score,
    precision_score,
    recall_score,
    f1_score
)
accuracy = accuracy_score(y_test, y_pred)
sensitivity = recall_score(y_test, y_pred, average='macro')
specificity = specificity_score(y_test, y_pred, average='macro')
precision = precision_score(y_test, y_pred, average='macro')
f1 = f1_score(y_test, y_pred, average='macro')


```
